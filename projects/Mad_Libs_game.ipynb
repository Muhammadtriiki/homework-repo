**Explanation of the Code below:**

**Function** choose_story():

* Prompts the user to select a story template.
* Asks for input (1 or 2) to choose between "Fairy Tale" and "Detective Story".

**Function** story_template_one():

* Returns a template for a Fairy Tale.
* he template includes placeholders for user inputs like:
   * {time_word}, {adjective}, {noun}, {place}, {verb}, {plural_noun},
      {creature}, {name}, etc.
* The placeholders represent the story's structure, where user-provided words
  will be inserted to create a unique story.

**Function** story_template_two():

* Returns a template for a Detective Story.
* Contains placeholders like:
  * {name}, {noun}, {number}, {time_period}, {adjective}, {place},
    {plural_noun}, {verb}, {object}, {clothing}, {adverb}, etc.
* These placeholders are used to fill in user-generated inputs to form the
  story.

**Function** collect_words(prompts):

* Prompts the user to enter values for each placeholder from the chosen story
  template.
* Takes a list of tuples (prompts) where each tuple consists of a key
  (placeholder) and a message.
* Loops through the prompts, asking for input and storing the responses in a
  dictionary (words).
* Tracks how many inputs have been provided and how many are left.

**Function** main():

* Main loop:
* Calls choose_story() to get userâ€™s story choice.
* Based on the choice, selects the appropriate story template and prompts.
* Calls collect_words(prompts) to gather user inputs for the selected story.
* Fills the story template with user-provided words using .format(**user_words).

**Story Templates and Prompts:**

* For Fairy Tale:
* time_word, adjective, noun, place, plural_noun, creature, name, etc.
* For Detective Story:
* name, noun, number, time_period, adjective, plural_noun, place, object  clothing, adverb, etc.

main() **loop continuation:**

* Asks if the user wants to play again.
* If the response is not "yes", exits the game.

```
def choose_story():
    print("Choose a story template:")
    print("1. Fairy Tale")
    print("2. Detective Story")
    return input("Enter 1 or 2: ").strip()

def story_template_one():
    return (
        "Once upon a {time_word}, there lived a {adjective} {noun} in a {adjective} {place}. "
        "Every {time_period}, they would {verb} to the nearby {place} to collect {plural_noun}. "
        "One day, they met a {adjective} {creature} who offered three {plural_noun}. "
        "The {noun} decided to {verb} the {plural_noun} with their {adjective} friend, {name}. "
        "Together, they used the {plural_noun} to create a magical {noun} that could {verb} {adverb}."
    )

def story_template_two():
    return (
        "Detective {name} had been working on the case of the missing {noun} for {number} {time_period}. "
        "The {adjective} investigator noticed a {adjective} {noun} at the crime scene, along with {number} {plural_noun}. "
        "Following the clues, they {verb} to the nearby {place}, where they discovered a {adjective} suspect {verb} behind a {noun}. "
        "The suspect was wearing a {adjective} {clothing} and holding a {object}. "
        "Using their {adjective} skills, Detective {name} solved the case while {verb} {adverb}."
    )

def collect_words(prompts):
    words = {}
    count = 0
    total_prompts = len(prompts)

    for key, message in prompts:
        print(f"\nYou have entered {count} words. {total_prompts - count} words left.")
        words[key] = input(message + ": ")
        count += 1

    return words

def main():
    while True:
        choice = choose_story()
        if choice == "1":
            story_template = story_template_one()
            prompts = [
                ("time_word", "Enter a time word (e.g., day, night)"),
                ("adjective", "Enter an adjective"),
                ("noun", "Enter a noun (character)"),
                ("adjective", "Enter another adjective"),
                ("place", "Enter a place"),
                ("time_period", "Enter a time period"),
                ("verb", "Enter a verb"),
                ("place", "Enter another place"),
                ("plural_noun", "Enter a plural noun"),
                ("adjective", "Enter another adjective"),
                ("creature", "Enter a magical creature"),
                ("plural_noun", "Enter another plural noun"),
                ("verb", "Enter another verb"),
                ("adjective", "Enter another adjective"),
                ("name", "Enter a name"),
                ("noun", "Enter another noun"),
                ("verb", "Enter another verb"),
                ("adverb", "Enter an adverb")
            ]
        elif choice == "2":
            story_template = story_template_two()
            prompts = [
                ("name", "Enter a detective name"),
                ("noun", "Enter a noun (missing item)"),
                ("number", "Enter a number"),
                ("time_period", "Enter a time period"),
                ("adjective", "Enter an adjective"),
                ("adjective", "Enter another adjective"),
                ("noun", "Enter a noun"),
                ("number", "Enter another number"),
                ("plural_noun", "Enter a plural noun"),
                ("verb", "Enter a verb (past tense)"),
                ("place", "Enter a place"),
                ("adjective", "Enter another adjective"),
                ("verb", "Enter a verb ending in -ing"),
                ("noun", "Enter a noun"),
                ("adjective", "Enter another adjective"),
                ("clothing", "Enter an article of clothing"),
                ("object", "Enter an object"),
                ("adjective", "Enter another adjective"),
                ("verb", "Enter a verb"),
                ("adverb", "Enter an adverb")
            ]
        else:
            print("Invalid choice. Please try again.")
            continue

        user_words = collect_words(prompts)
        story = story_template.format(**user_words)
        print("\nHere is your story:\n")
        print(story)

        while True:
            replay = input("\nWould you like to play again? (yes/no): ").strip().lower()
            if replay in ("yes", "no" , "y" , "n"):
                break
            else:
                print("Invalid answer. Please enter 'yes' or 'no'.")
        if replay != "yes":
            print("Thanks for playing!")
            break

if __name__ == "__main__":
    main()
```
